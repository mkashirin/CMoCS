### Задание 

Построить множество точек, соответствующих следующим фигурам:

__Фигуры Лиссажу__
$$ \begin{equation} \left\{ \begin{array}{}{} x = A_x \textrm{sin}(\omega_x t); &\\ y = A_x \textrm{sin}(\omega_x t). \end{array} \right. (1) \end{equation} $$

__Эпитрохойда__
$$ \begin{equation} \left\{ \begin{array}{}{} x = (R + r)\textrm{cos}(\frac{r}{R}\phi) - h\textrm{cos}(\frac{R +r}{r}\phi); &\\ y = (R + r)\textrm{sin}(\frac{r}{R}\phi) - h\textrm{sin}(\frac{R +r }{r}\phi). \end{array} \right. (2) \end{equation} $$

__Гипотрохойда__
$$ \begin{equation} \left\{ \begin{array}{}{} x = (R - r)\textrm{cos}\phi + h\textrm{cos}(\frac{R +r}{r}\phi); &\\ y = (R - r)\textrm{sin}\phi - h\textrm{sin}(\frac{R + r}{r}\phi). \end{array} \right. (3) \end{equation}$$

Во всех вышеописааных системах $\phi$ изменяется от $0$ до $2\pi$.

Для получения траектории движения колеса, необходимо изменять
значение $\phi$ от $0$ до $30$, с шагом $0.01$ вычислять координаты и представлять их
на графике.

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Функция для фигур Лиссажу

Напишем функцию для построения фигур Лиссажу с различными параметрами.

```python
def plot_lissajous_curve(
        big_a_x: float, big_a_y: float, omega_x: float, 
        omega_y: float, t_0: float, t_1: float,
        delta: float
):
    # Both x and y coordinates should be stored in separate lists to
    # have the ability to be plotted correctly
    x_points, y_points = list(), list()
    
    for t in np.arange(t_0, t_1, delta):
        x = big_a_x * np.sin(omega_x * t)
        x_points.append(x)
        y = big_a_y * np.sin(omega_y * t)
        y_points.append(y)

    # Passing x and y to the plt.plot() function as two separate arrays
    plt.plot(x_points, y_points)
    plt.title("Lissajous curve")
    plt.axis("equal")
    plt.show()

    # Converting coordinates to a structured numpy array 
    points = np.array(
        [x_points, y_points],
        dtype=[("x_points", float), ("y_points", float)]
    )
    return points
```

```python
plot_lissajous_curve(
    big_a_x=1, big_a_y=1, omega_x=3, 
    omega_y=4, t_0=0, t_1=2*np.pi,
    delta=.01
)
```

#### Функция для Эпитрохойды

Напишем функцию для построения эпитрохойды с различными параметрами.

```python
def plot_epitrochoid(
        big_r: float, small_r: float, phi_0: float, 
        phi_1: float, fac: float, delta: float
):
    # Both x and y coordinates should be stored in separate lists to
    # have the ability to be plotted correctly
    x_points, y_points = list(), list()

    for phi in np.arange(phi_0, phi_1, delta):
        x = ((big_r + small_r) * np.cos(small_r / big_r * phi) -
             fac * np.cos(((big_r + small_r) / big_r) * phi))
        x_points.append(x)
        y = ((big_r + small_r) * np.sin(small_r / big_r * phi) -
             fac * np.sin(((big_r + small_r) / big_r) * phi))
        y_points.append(y)

    # Passing x and y to the plt.plot() function as two separate arrays
    plt.plot(x_points, y_points)
    plt.title("Epitrochoid")
    plt.axis("equal")
    plt.show()

    # Converting coordinates to a structured numpy array 
    points = np.array(
        [x_points, y_points],
        dtype=[("x_points", float), ("y_points", float)]
    )
    return points
```

```python
plot_epitrochoid(
    big_r=5, small_r=3, phi_0=0,
    phi_1=30*np.pi, fac=7, delta=.01
)
```

#### Функция для Гипотрохойды

Напишем функцию для построения гипотрохойды с различными параметрами.

```python
def plot_hypotrochoid(
        big_r: float, small_r: float, phi_0: float,
        phi_1: float, fac: float, delta: float
):
    # Both x and y coordinates should be stored in separate lists to
    # have the ability to be plotted correctly
    x_points, y_points = list(), list()

    for phi in np.arange(phi_0, phi_1, delta):
        x = ((big_r - small_r) * np.cos(phi) +
             fac * np.cos(((big_r - small_r) / big_r) * phi))
        x_points.append(x)
        y = ((big_r - small_r) * np.sin(phi) -
             fac * np.sin(((big_r - small_r) / big_r) * phi))
        y_points.append(y)

    # Passing x and y to the plt.plot() function as two separate arrays
    plt.plot(x_points, y_points)
    plt.title("Hypotrochoid")
    plt.axis("equal")
    plt.show()

    # Converting coordinates to a structured numpy array 
    points = np.array(
        [x_points, y_points],
        dtype=[("x_points", float), ("y_points", float)]
    )
    return points
```

```python
plot_hypotrochoid(
    big_r=5, small_r=3, phi_0=0,
    phi_1=30*np.pi, fac=5/3, delta=.01
)
```
