### Задание

В колесе радиуса $R$ катится колесо радиуса $r$. Какую траекторию описывает точка, расположенная на ободе колеса $r$?

Траектория движения этой точки находится при помощи системы (1).

$$ \begin{equation} \left\{ \begin{array}{}{} x = (R - r)\textrm{cos}\phi + d\textrm{cos}(\phi - \frac{r}{R}\phi) &\\ y = (R - r)\textrm{sin}\phi + d\textrm{sin}(\phi - \frac{r}{R}\phi)\end{array} \right. (1) \end{equation} $$

В этой системе $\phi$ изменяется от $0$ до $2\pi$ (угол смещения колеса $r$). Для получения траектории движения колеса, необходимо изменять значение $\phi$ от $0$ до $30$, с шагом $0.01$, вычислять координаты и представлять их на графике.

#### Функция описания траектории точки

Реализуем функцию, которая будет отображать движение точки по траектории с различными параметрами.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_trajectory(
        big_r: float, small_r, phi_0: float,
        phi_1: float, delta: float
):
    # Both x and y coordinates should be stored in separate lists to
    # have the ability to be plotted correctly
    x_points, y_points = list(), list()
    
    for phi in np.arange(phi_0, phi_1, delta):
        x = ((big_r - small_r) * np.cos(phi) +
             small_r * np.cos(phi - (small_r / big_r) * phi))
        x_points.append(x)
        y = ((big_r - small_r) * np.sin(phi) +
             small_r * np.sin(phi - (small_r / big_r) * phi))
        y_points.append(y)
        
    # Passing x and y to the plt.plot() function as two separate arrays
    plt.plot(x_points, y_points)
    plt.title("Point trajectory")
    plt.axis("equal")
    plt.show()
    
    # Converting coordinates to a structured numpy array 
    points = np.array(
        [x_points, y_points],
        dtype=[("x_points", float), ("y_points", float)]
    )
    return points
```

```python
plot_trajectory(
    big_r=3, small_r=1, phi_0=0, 
    phi_1=30*np.pi, delta=.01
)
```
