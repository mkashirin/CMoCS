#### Задача 1

Построить модель затухающих гармонических колебаний, для которых колебательный процесс описывается уравнениями, где $ \delta $ — коэффициент затухания. Самые уравнения представлены ниже.

$$ \begin{equation} y = Ae^{-\delta t_n} \textrm{sin} \phi \end{equation} $$
$$ \begin{equation} \phi = \omega t_n + \phi_0 \end{equation} $$
$$ \begin{equation} \omega = \sqrt{\omega_0^2 - \delta^2} \end{equation} $$

Дискретное время вычислить по формуле (4), где $n = 0, 1, 2, ...$, $T_0$ — период дискретизации модели. Расчёты провести в соответствии с вариантом, представленным в таблице.
&nbsp;

| № варианта  |  $T_0$  |  $A$  |  $\omega$  |  $\phi_0$  | $\delta$ |
|:-----------:|:-------:|:-----:|:----------:|:----------:|:--------:|
|      1      |  0.001  |   1   |    0.2     |     0      |   0.01   |
|      2      |  0.001  |  0.9  |    0.25    |    0.05    |   0.02   |
|      3      |  0.001  |  0.8  |    0.3     |    0.1     |   0.03   |
|      4      |  0.001  |  0.7  |    0.35    |    0.15    |   0.04   |
|      5      |  0.001  |  0.6  |    0.4     |    0.2     |   0.05   |
|      6      |  0.01   |   1   |    0.2     |     0      |   0.06   |
|      7      |  0.01   |  0.9  |    0.25    |    0.05    |   0.07   |
|      8      |  0.01   |  0.8  |    0.3     |    0.1     |   0.08   |
|      9      |  0.01   |  0.7  |    0.35    |    0.15    |   0.09   |
|     10      |  0.01   |  0.6  |    0.4     |    0.2     |   0.1    |

```python
def get_model(
        points_number: int, 
        big_t: float, big_a: float, phi_0: float, omega_0: float, delta: float
):
    # Both timestamps and computed should be stored in separate lists to
    # have the ability to be plotted correctly
    timestamps, computed = list(), list()
    for t in np.arange(big_t, points_number, big_t):
        # Performing computations
        timestamps.append(t)
        omega = np.sqrt(omega_0 ** 2 - delta ** 2)
        phi = omega * t + phi_0
        y = big_a * np.e ** (-delta * t) * np.sin(phi)
        computed.append(y)
    return timestamps, computed
```

Построить график полученной модели согласно набору точек равному 1000 измерений.

##### Примечание

_Значение амплитуды увеличим в 500 раз, а начального угла — в 10. Так изменения убывающего осциллятора станут заметнее._

```python
# Completing the task for variant 9
measurements, y_array = get_model(
    points_number=1000, 
    big_t=0.01, big_a=0.7 * 500, phi_0=0.15 * 10, omega_0=0.35, delta=0.01
)

plt.plot(measurements, y_array)
plt.title("Damped oscillations")
plt.axis("equal")
plt.show()
```
